Data Exploration and Wrangling

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/crumm/OneDrive/Documents/MSDS Master Folder/MSDS Fall 2025/New_York_City_Leading_Causes_of_Death_20251103.csv")
print(df.head())
df.info() # we can see that the year is an integer, however the columns for deaths and death rates are considered "objects"
#in order to get their summary statistics we must change their column type

cols = ['Deaths', 'Death Rate', 'Age Adjusted Death Rate']
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')

df.dtypes # we now have all necessary columns as integers

summary_df = df.describe()
print(summary_df) # here we can see the summary statistics of our columns, however the most valuable to us is likely those of our "Deaths" column

print(df.isnull().sum()) # here we can see that there are 138 missing values for deaths, and 729 missing values for death rates and age adjusted death rates each

df = df.rename(columns={
    'Leading Cause': 'Cause_of_Death',
    'Race Ethnicity': 'Race_Ethnicity',
    'Death Rate': 'Death_Rate',
    'Age Adjusted Death Rate': 'Age_Adjusted_Death_Rate'
    })
```

Visualizations

```python
plt.figure(figsize=(10,6)) 
plt.bar(df['Cause_of_Death'], df['Deaths'], color='red') # bar plot of number of deaths by cause of death
plt.xticks(rotation=90, fontsize=6) # changed font size of causes of death for legibility
plt.xlabel('Cause of Death', fontsize=14) # changed font size of x axis label for legibility
plt.ylabel('Number of Deaths', fontsize=14) # changed font size of y axis label for consistency
plt.title('Number of Deaths by Cause of Death', fontsize=18) # changed font size of title for legibility
plt.show()
```

Here we can see a graph demonstrating the number of deaths for each cause of death listed in our data. While not entirely practical/legible considering the very long values of "Cause of Death" we are still able to Covid, heart disease and cancer were big killers according to the data.

```python
plt.figure(figsize=(10,6))
brs = plt.bar(df['Race_Ethnicity'], df['Deaths'], color='orange', label='Deaths') # demonstrating ability to change the color of our graph
plt.xticks(rotation=45, fontsize=10, ha='right') # demonstrating how to adjust horizontal alignment of x axis labels
plt.xlabel('Race/Ethnicity', fontsize=14)
plt.ylabel('Number of Deaths', fontsize=14)
plt.title('Number of Deaths by Race Ethnicity', fontsize=18)
plt.legend() # added a legend, and while it does not demonstrate anything new, it does demonstrate the ability to add a legend!
plt.show()
```

In this graph we can see the number of deaths by Race/Ethnicity. In this graph, I have continued to show how to adjust our labels, titles, values and colors within the graph.
```python
import seaborn as sns
plt.figure(figsize=(10,6))
sns.barplot(x='Cause_of_Death', y='Deaths', data=df, color='red')
plt.xticks(rotation=90, fontsize=6)
plt.xlabel('Cause of Death', fontsize=14)
plt.ylabel('Number of Deaths', fontsize=14)
plt.title('Distribution of Deaths by Cause of Death', fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Race_Ethnicity', y='Deaths', data=df, color='orange')
plt.xticks(rotation=45, fontsize=10, ha='right')
plt.xlabel('Race Ethnicity', fontsize=14)
plt.ylabel('Number of Deaths', fontsize=14)
plt.title('Number of Deaths by Race Ethnicity', fontsize=18)
plt.legend(['Deaths'])
plt.show()
```
